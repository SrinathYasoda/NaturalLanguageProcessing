```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('G:\\JOBIFY\\archive\\IMDB_Dataset.csv')
data.head(20)
```

```python
df=data
```

```python
pt=df[df['sentiment']=='positive']
pt=pt['review']
```

```python
#df.append(pt)
```

```python
nt=df[df['sentiment']=='negative']
nt=nt['review']
```

```python
#plotting the word cloud
allwordspt=' '.join(twt for twt in pt)
allwordsnt=' '.join(twt for twt in nt)
```

```python
wordCloud=WordCloud(width=600,height=300,random_state=21,max_font_size=100,background_color='white',stopwords=set(STOPWORDS)).generate(allwordspt)

#plotting
plt.imshow(wordCloud,interpolation='bilinear')
plt.axis('off')
plt.show

```

```python
#Negative WordCloud
wordCloud=WordCloud(width=600,height=300,random_state=21,max_font_size=100,background_color='white',stopwords=set(STOPWORDS)).generate(allwordsnt)

#plotting
plt.imshow(wordCloud,interpolation='bilinear')
plt.title("Negative Word Cloud")
plt.axis('off')
plt.show

```
